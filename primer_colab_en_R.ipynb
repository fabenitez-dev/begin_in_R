# Ejemplo de R en google colab

Vamos a simular $n=50$ observaciones de una $N(\mu=170, \sigma=10)$ que vamos a almacenar en el vector `estatura` y luego vamos a construir un histograma


```r
estatura <- rnorm(n=50, mean=170, sd=10)
hist(estatura, col='blue', xlab='Estatura (cm)', ylab='Frecuencia', main='Histograma de estaturas')
```
